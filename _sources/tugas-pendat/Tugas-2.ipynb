```python
# Import Library yang diperlukan
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
# Membaca dan Menampilkan data csv
iris = pd.read_csv('https://raw.githubusercontent.com/rifqi-522/datamining/main/Iris.csv')
iris.head()
```

```python
# mengetahui jumlah baris dan kolom
iris.shape
```

```python
# proses diskritisasi sepal.length
from sklearn.preprocessing import KBinsDiscretizer
a = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
iris["SepalLengthCm"] = a.fit_transform(iris[["SepalLengthCm"]])
```

```python
# menampilkan hasil diskritisasi
iris["SepalLengthCm"]
```

```python
# proses diskritisasi sepal.width
from sklearn.preprocessing import KBinsDiscretizer
a = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
iris["SepalWidthCm"] = a.fit_transform(iris[["SepalWidthCm"]])
```

```python
# menampilkan hasil diskritisasi
iris["SepalWidthCm"]
```

```python
# proses diskritisasi petal.length
from sklearn.preprocessing import KBinsDiscretizer
a = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
iris["PetalLengthCm"] = a.fit_transform(iris[["PetalLengthCm"]])
```

```python
# menampilkan hasil diskritisasi
iris["PetalLengthCm"]
```

```python
# proses diskritisasi petal.width
from sklearn.preprocessing import KBinsDiscretizer
a = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
iris["PetalWidthCm"] = a.fit_transform(iris[["PetalWidthCm"]])
```

```python
# menampilkan hasil diskritisasi
iris["PetalWidthCm"]
```
