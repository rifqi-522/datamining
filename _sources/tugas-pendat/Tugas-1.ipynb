```python
# import library
import numpy as np
import pandas as pd
import seaborn as sb
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
#menampilkan data
data = pd.read_csv("https://raw.githubusercontent.com/rifqi-522/datamining/main/car_price_prediction.csv")
print(data)
```

```python
#menampilkan data  5 pertama dan data 5 terakhir
data
```

```python
# 1. 
def dt1(i,j):
  P = 9
  Q = 0
  for col in ["Manufacturer", "Model", "Prod. year", "Category", "Fuel type", "Gear box type", "Doors", "Wheel", "Color"]:
    if data[col][i] == data[col][j]:
      Q += 1
  return (P-Q)/P
```

```python
print(dt1(1,2))
print(dt1(1,3))
print(dt1(1,4))
```

```python
# Numerik
def dt2(i,j,h):
  result = 0
  for col in ["Price", "Cylinders", "Airbags"]:
    result = math.pow(abs(data[col][i] - data[col][j]), h)
  return result**(1/h)
```

```python
print(dt2(1,2,1))
print(dt2(1,3,1))
print(dt2(1,4,1))
```

```python
print(dt2(1,2,2))
print(dt2(1,3,2))
print(dt2(1,4,2))
```
